```python
minus_review = []
for elem in bad_review:
    minus_review.append(elem.text)
#список отрицательных отзывов (30 штук)
```

```python
review_dict = {}
for text in plus_review:
    review_dict[text] = '+'
for some in minus_review:
    review_dict[some] = '-'

review_dict
```

```python
import nltk
from nltk.tokenize import word_tokenize
```

```python
from string import punctuation
from pymorphy2 import MorphAnalyzer
morph = MorphAnalyzer()
```

```python
#токенизируем рецензии, приводим все слова к нижнему регистру, лемматизируем, 
#создаем отдельный список для слов из положительных отзывав, отдельный - для слов из отрицательных отзывов
plus_words_list = []
minus_words_list = []
for elem in review_dict.keys():
    if review_dict[elem] == '+':
        tokenizer = word_tokenize(elem)
        for some in tokenizer:
            if some not in punctuation:
                if some.isalpha:
                    some = some.lower()
                    some = morph.parse(some)[0].normal_form
                    plus_words_list.append(some)
    elif review_dict[elem] == '-':
        tokenizer2 = word_tokenize(elem)
        for word in tokenizer2:
            if word not in punctuation:
                if word.isalpha:
                    word = word.lower()
                    word = morph.parse(word)[0].normal_form
                    minus_words_list.append(word)
print (plus_words_list[0:10])
print (minus_words_list[0:10])
```

```python
plus_words = [] #создаем новые списки для неповторяющихся слов (чтобы не удалять повторяющиеся элементы из старых и тем самым 
#не нарушать сравнение)
minus_words = []
for elem in plus_words_list:
    if elem not in minus_words_list:
        plus_words.append(elem)
for word in minus_words_list:
    if word not in plus_words_list:
        minus_words.append(word)
```

```python
from collections import Counter
```

```python
cnt_plus = Counter(plus_words)
plus = []
cnt_minus = Counter (minus_words)
minus = []
for r in cnt_plus.keys():
    if cnt_plus[r] > 1:
        plus.append(r)
for n in cnt_minus.keys():
    if cnt_minus[n]>1:
        minus.append(n)
plus = set(plus) #множество частотных слов из положительных отзывов (слова частотностью 1 удалены)
minus = set(minus) #множество частотных слов из отрицательных отзывов (слова частотностью 1 удалены)
print (plus)
print (minus)
```

Импортируем тексты отзывов для тестирования программы: тексты отзывов токенизированы, лемматизированы и приведены к нижнему регистру. Взяты 5 положительных, 5 отрицательых отзывов

```python
from reviews import first_minus, second_minus, third_minus, forth_minus, fifth_minus, first_plus, second_plus, third_plus, forth_plus, fifth_plus
```

Функция для определения тональности:

```python
def review_tone (review_text):
    counter = {'plus_counter': 0, 'minus_counter': 0}
    for word in review_text:
        if word in plus:
            counter['plus_counter'] += 1
        if word in minus:
            counter ['minus_counter'] +=1
    if counter ['plus_counter'] > counter ['minus_counter']:
        print ('+')
    elif counter ['plus_counter'] < counter ['minus_counter']:
        print ('-')
    else:
        print ('отзыв написан имеет нейтральную тональность')
```

```python
review_tone (first_plus)
review_tone (second_plus)
review_tone (third_plus)
review_tone (forth_plus)
review_tone (fifth_plus) #все положительные отзывы в итоге определились как положительные
```

```python
review_tone (first_minus)
review_tone (second_minus)
review_tone (third_minus)
review_tone (forth_minus)
review_tone (fifth_minus) #к сожалению, все отрицательные отзывы тоже определились как положительные
```

```python
text = ['школьный', 'дикий', 'простой']
review_tone (text) #данный пример доказывает, что функция все-таки может определить отрицательную тональность, но на текстах отзывов она не сработала
```
